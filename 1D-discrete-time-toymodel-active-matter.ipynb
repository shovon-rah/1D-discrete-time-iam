```python
import numpy as np
import random
import matplotlib.pyplot as plt

random.seed(4)

N = 300
L = 100
T = 10000
xmin, xmax = -L, L

pos = [random.randint(xmin, xmax) for _ in range(N)]

plt.figure(figsize=(6,4))
plt.hist(pos, bins=50, density=True)
plt.axvline(0, color='k', linestyle='--')
plt.xlabel("Position")
plt.ylabel("Probability density")
plt.title("Initial position distribution")
plt.show()

print("Left room:", sum(1 for x in pos if x < 0))
print("Right room:", sum(1 for x in pos if x > 0))
```

```python
for t in range(T):
    new_pos = []
    
    for x in pos:
        hop = random.choice([-1, 1])
        x_trial = x+ hop

        # reflecting boundary conditions
        if x_trial < xmin:
            x_trial = x+1
        elif x_trial > xmax:
            x_trial = x-1

        if x < 0 and x_trial >= 0:
            x_trial = x   # reflect

        new_pos.append(x_trial)
    pos = new_pos

plt.figure(figsize=(6,4))
plt.hist(pos, bins=50, density=True)
plt.axvline(0, color='k', linestyle='--')
plt.xlabel("Position")
plt.ylabel("Probability density")
plt.title("Final position distribution")
plt.show()

print("Left room:", sum(1 for x in pos if x < 0))
print("Right room:", sum(1 for x in pos if x > 0))
```
